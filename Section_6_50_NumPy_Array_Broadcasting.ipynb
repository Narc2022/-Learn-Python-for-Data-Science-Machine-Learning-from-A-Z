#broadcasting
The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations.
Screenshot 2023-04-23 at 7.04.13 PM.png

```python
import numpy as np

a = np.array([1,2,3,4,5])
b = np.array([10,0,20,1,2])
c = a*b
print('1 ',c)
print ('\n') 
print('-----------------------------------')
print ('\n') 


import numpy as np 
a = np.array([[0.0,0.0,0.0],[10.0,10.0,10.0],[20.0,20.0,20.0],[30.0,30.0,30.0]]) 
b = np.array([1.0,2.0,3.0])  
   
print('First array:') 
print (a) 
print ('\n')  
   
print ('Second array:') 
print (b) 
print ('\n')  
   
print ('2 ','First Array + Second Array')
print(a + b)
```

#Comparison functions



```python
import numpy as np

x = np.array([1,2,3,4,5,6])

print('x<3  :',x<3)
print('x>3  :',x>3)
print('x<=3 :',x<=3)
print('x>=3 :',x>=3)
print('x!=3 :',x!=3)
print('x==3 :',x==3)
 
```
